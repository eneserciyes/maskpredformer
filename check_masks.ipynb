```python
import matplotlib.pyplot as plt
import torch
import os
from torchmetrics import JaccardIndex
```

```python
jaccard = JaccardIndex(task="multiclass", num_classes=49, average='macro')
```

```python
data_root = "data_prepared/train/video_0"
```

```python
masks = torch.load(os.path.join(data_root, "mask.pt"))
gt_masks = torch.load(os.path.join(data_root, "mask_gt.pt"))
```

```python
plt.imshow(torch.argmax(masks[0], dim=0))
```

```python
plt.imshow(gt_masks[0])
```

```python
jaccard(torch.argmax(masks[0], dim=0), gt_masks[0])
```
